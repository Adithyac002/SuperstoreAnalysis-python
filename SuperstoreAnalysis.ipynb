```python
# Check the number of rows and columns
print("Rows and Columns:", df.shape)

# Display the column names and their data types
print("\nColumn Names and Data Types:")
print(df.dtypes)

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
total_sales=df['Sales'].sum()
total_profit=df['Profit'].sum()
Average_discount=df['Discount'].mean()
region_profit=df.groupby('Region')['Profit'].sum()
highest_region_profit=region_profit.idxmax()
print("Total Sales: ",total_sales)
print("Total_Profit: ",total_profit)
print("Average_discount: ", Average_discount)
print("Region Profit: ",region_profit)
print("Highest Region Profit: ",highest_region_profit)
```

```python
top_sales= df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(3)
print("Top 3 Sub_Category by Sales")
print(top_sales)
```

```python
bottom_profit=df.groupby('Sub-Category')['Sales'].sum().sort_values().head(3)
print("\nBottom 3 Sub-Categories by Profit:")
print(bottom_profit)
```

```python
df['Order Date']=  pd.to_datetime(df['Order Date'])
df['Month_Year']= df['Order Date'].dt.to_period('M')
monthly_sales=df.groupby('Month_Year')['Sales'].sum()
monthly_sales.plot(kind='line',figsize=(12,6),marker='o',color='green')
plt.title("Sales Over Time")
plt.xlabel('Month_Year')
plt.ylabel('Total_Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_customers=df.groupby('Customer Name')[['Sales','Profit']].sum()
top_5=top_customers.sort_values(by='Sales', ascending=False).head(5)
print(top_5)
```

```python
top_5.sort_values(by='Sales').plot(kind='barh', figsize=(10,6), color=['skyblue', 'lightgreen'])
plt.title("Top 5 Customers by Sales and Profit")
plt.xlabel("Amount")
plt.ylabel("Customer Name")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
pivot=df.pivot_table(values='Profit', index='Segment',columns='Region', aggfunc='sum')
print("Profit by Segment and Region: ")
print(pivot)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Profit Distribution by Segment and Region")
plt.ylabel("Segment")
plt.xlabel("Region")
plt.tight_layout()
plt.show()
```
